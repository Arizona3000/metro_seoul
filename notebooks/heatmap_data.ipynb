```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('flows.csv')
```

```python
data.head()
```

```python
# 1) I filter for data of one single day
new_data = data[data['date'] == '2023-06-30']
```

```python
# 2) I delete columns that are not interesting for the plot
new_data.drop(columns=['line', 'station_number', 'Unnamed: 0', 'date'], inplace=True)
```

```python
new_data.head()
```

```python
# 3) I group by station_names
new_data = new_data.groupby(['station_name']).sum().reset_index()
new_data.head()
```

```python
# 4) I drop data for the last column '24', given it's confusing data
new_data = new_data.drop(columns=['24'])
```

```python
# 5) I create a list with column names and transpose the df

column_names = list(new_data.columns[2:])
new_data = pd.melt(new_data, id_vars=['station_name'], value_vars=column_names)
new_data.head(10)
```

```python
# 6) I rename the columns

new_data.rename(columns={'variable': 'hour', 'value': 'n_passengers'}, inplace=True)
new_data.sort_values(by=['station_name'])
new_data.head()
```

```python
new_data.value_counts('station_name')
```

## Add location data and translation data


```python
# 1) I upload and clean  the translation
map = pd.read_excel('translate2.xlsx')
map.drop(map.columns[0:2], axis=1, inplace=True)
map.rename(columns={map.columns[1]: 'station_name', map.columns[0]: 'korean_name'}, inplace=True)
map.head(), map.shape
```

```python
# 1.1) Checking for the nature duplicates before dropping them
dupl_korean = map[map['korean_name'].duplicated()].index.values
dupl_english = map[map['station_name'].duplicated()].index.values
diff = [el for el in dupl_korean if el not in dupl_english]
diff
```

```python
map.drop_duplicates(inplace=True)
```

```python
# 2) I upload and clean location data
loc = pd.read_csv('station_gps.csv')
loc.drop(columns=['Unnamed: 0', 'line', 'no'], inplace=True)
loc.rename(columns={'name': 'korean_name'}, inplace=True)
loc.head(), loc.shape
```

```python
loc.drop_duplicates(inplace=True)
```

```python
# 3) I merge the location data with translation data

merged = map.merge(loc, on='korean_name')
merged.head(), merged.shape
```

```python
merged.head()
```

### Merging the new df with location data with the data on passenger flow

```python
import plotly.graph_objects as go

import pandas as pd

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
dataset = pd.read_csv(url)

years = ["1952", "1962", "1967", "1972", "1977", "1982", "1987", "1992", "1997", "2002",
         "2007"]

# make list of continents
continents = []
for continent in dataset["continent"]:
    if continent not in continents:
        continents.append(continent)
# make figure
fig_dict = {
    "data": [],
    "layout": {},
    "frames": []
}

# fill in most of layout
fig_dict["layout"]["xaxis"] = {"range": [30, 85], "title": "Life Expectancy"}
fig_dict["layout"]["yaxis"] = {"title": "GDP per Capita", "type": "log"}
fig_dict["layout"]["hovermode"] = "closest"
fig_dict["layout"]["updatemenus"] = [
    {
        "buttons": [
            {
                "args": [None, {"frame": {"duration": 500, "redraw": False},
                                "fromcurrent": True, "transition": {"duration": 300,
                                                                    "easing": "quadratic-in-out"}}],
                "label": "Play",
                "method": "animate"
            },
            {
                "args": [[None], {"frame": {"duration": 0, "redraw": False},
                                  "mode": "immediate",
                                  "transition": {"duration": 0}}],
                "label": "Pause",
                "method": "animate"
            }
        ],
        "direction": "left",
        "pad": {"r": 10, "t": 87},
        "showactive": False,
        "type": "buttons",
        "x": 0.1,
        "xanchor": "right",
        "y": 0,
        "yanchor": "top"
    }
]

sliders_dict = {
    "active": 0,
    "yanchor": "top",
    "xanchor": "left",
    "currentvalue": {
        "font": {"size": 20},
        "prefix": "Year:",
        "visible": True,
        "xanchor": "right"
    },
    "transition": {"duration": 300, "easing": "cubic-in-out"},
    "pad": {"b": 10, "t": 50},
    "len": 0.9,
    "x": 0.1,
    "y": 0,
    "steps": []
}

# make data
year = 1952
for continent in continents:
    dataset_by_year = dataset[dataset["year"] == year]
    dataset_by_year_and_cont = dataset_by_year[
        dataset_by_year["continent"] == continent]

    data_dict = {
        "x": list(dataset_by_year_and_cont["lifeExp"]),
        "y": list(dataset_by_year_and_cont["gdpPercap"]),
        "mode": "markers",
        "text": list(dataset_by_year_and_cont["country"]),
        "marker": {
            "sizemode": "area",
            "sizeref": 200000,
            "size": list(dataset_by_year_and_cont["pop"])
        },
        "name": continent
    }
    fig_dict["data"].append(data_dict)

# make frames
for year in years:
    frame = {"data": [], "name": str(year)}
    for continent in continents:
        dataset_by_year = dataset[dataset["year"] == int(year)]
        dataset_by_year_and_cont = dataset_by_year[
            dataset_by_year["continent"] == continent]

        data_dict = {
            "x": list(dataset_by_year_and_cont["lifeExp"]),
            "y": list(dataset_by_year_and_cont["gdpPercap"]),
            "mode": "markers",
            "text": list(dataset_by_year_and_cont["country"]),
            "marker": {
                "sizemode": "area",
                "sizeref": 200000,
                "size": list(dataset_by_year_and_cont["pop"])
            },
            "name": continent
        }
        frame["data"].append(data_dict)

    fig_dict["frames"].append(frame)
    slider_step = {"args": [
        [year],
        {"frame": {"duration": 300, "redraw": False},
         "mode": "immediate",
         "transition": {"duration": 300}}
    ],
        "label": year,
        "method": "animate"}
    sliders_dict["steps"].append(slider_step)


fig_dict["layout"]["sliders"] = [sliders_dict]

fig = go.Figure(fig_dict)

fig.show()
```
