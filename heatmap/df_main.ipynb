```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('flows.csv')
```

```python
data.head()
```

```python
# 1) I filter for data of one single day
new_data = data[data['date'] == '2023-06-30']
```

```python
# 2) I delete columns that are not interesting for the plot
new_data.drop(columns=['line', 'station_number', 'Unnamed: 0', 'date'], inplace=True)
```

```python
new_data.head()
```

```python
# 3) I group by station_names
new_data = new_data.groupby(['station_name']).sum().reset_index()
new_data.head()
```

```python
# 4) I drop data for the last column '24', given it's confusing data
new_data = new_data.drop(columns=['24'])
```

```python
# 5) I create a list with column names and transpose the df

column_names = list(new_data.columns[2:])
new_data = pd.melt(new_data, id_vars=['station_name'], value_vars=column_names)
new_data.head(10)
```

```python
# 6) I rename the columns

new_data.rename(columns={'variable': 'hour', 'value': 'n_passengers'}, inplace=True)
new_data.sort_values(by=['station_name'])
new_data.head()
```

```python
new_data.value_counts('station_name')
```

## Add location data and translation data


```python
# 1) I upload and clean  the translation
map = pd.read_excel('translate2.xlsx')
map.drop(map.columns[0:2], axis=1, inplace=True)
map.rename(columns={map.columns[1]: 'station_name', map.columns[0]: 'korean_name'}, inplace=True)
map.head(), map.shape
```

```python
# 1.1) Checking for the nature duplicates before dropping them
dupl_korean = map[map['korean_name'].duplicated()].index.values
dupl_english = map[map['station_name'].duplicated()].index.values
diff = [el for el in dupl_korean if el not in dupl_english]
diff
```

```python
map.drop_duplicates(inplace=True)
```

```python
# 2) I upload and clean location data
loc = pd.read_csv('station_gps.csv')
loc.drop(columns=['Unnamed: 0', 'line', 'no'], inplace=True)
loc.rename(columns={'name': 'korean_name'}, inplace=True)
loc.head(), loc.shape
```

```python
loc.drop_duplicates(inplace=True)
```

```python
# 3) I merge the location data with translation data

merged = map.merge(loc, on='korean_name')
merged.head(), merged.shape
```

```python
merged.head()
```

### Merging the new df with location data with the data on passenger flow

```python
# 4) I merge
unified = new_data.merge(merged, on='station_name')
unified.head()
```

```python
# We lose 20 information on metro stations due to this merge (probably english names not matching)
len(unified['station_name'].unique())
```

## Plot the on the city 

```python
img = px.scatter_mapbox(unified, lat="lat", lon="lng",
                        hover_name= "station_name",
                        color='n_passengers',
                        size = 'n_passengers',
                     animation_frame="hour", zoom=9,
                     color_continuous_scale=px.colors.cyclical.IceFire)
img.update_layout(mapbox_style="carto-positron")

img.show()
```

## Two separate plots for entry and exit

```python
df = pd.read_csv('flows.csv')
new_df = df[df['date'] == '2023-06-30']
new_df.drop(columns=['line', 'station_number', 'Unnamed: 0', 'date'], inplace=True)
new_df.head()
```

```python
new_df = new_df.groupby(['station_name', 'entry/exit']).sum().reset_index()
new_df.head()
```

```python
# 1) I split the df in two: entries and exits

df_entry = new_df[new_df['entry/exit'] == 'entry']
df_exit = new_df[new_df['entry/exit'] == 'exit']
```

```python
# 2.a) Preprocessing for df_entry
df_entry = df_entry.drop(columns=['24'])

column_names = list(df_entry.columns[2:])
df_entry = pd.melt(df_entry, id_vars=['station_name'], value_vars=column_names)
df_entry.head(10)

df_entry.rename(columns={'variable': 'hour', 'value': 'n_passengers_entry'}, inplace=True)
df_entry.sort_values(by=['station_name'])
df_entry.head()
```

```python
# 2.b) Preprocessing for df_exit
df_exit = df_exit.drop(columns=['24'])

column_names = list(df_exit.columns[2:])
df_exit = pd.melt(df_exit, id_vars=['station_name'], value_vars=column_names)
df_exit.head(10)

df_exit.rename(columns={'variable': 'hour', 'value': 'n_passengers_exit'}, inplace=True)
df_exit.sort_values(by=['station_name'])
df_exit.head()
```

### I merge the two datasets with 

```python
unified_entry = df_entry.merge(merged, on='station_name')
unified_entry.head()
```

```python
unified_exit = df_exit.merge(merged, on='station_name')
unified_exit.head()
```

```python
img = px.scatter_mapbox(unified_entry, lat="lat", lon="lng",
                        hover_name= "station_name",
                        color='n_passengers_entry',
                        size = 'n_passengers_entry',
                     animation_frame="hour", zoom=9,
                     color_continuous_scale=px.colors.cyclical.IceFire)
img.update_layout(mapbox_style="carto-positron")

img.show()
```

```python
img = px.scatter_mapbox(unified_exit, lat="lat", lon="lng",
                        hover_name= "station_name",
                        color='n_passengers_exit',
                        size = 'n_passengers_exit',
                     animation_frame="hour", zoom=9,
                     color_continuous_scale=px.colors.cyclical.IceFire)
img.update_layout(mapbox_style="carto-positron")

img.show()
```

```python
from plotly.subplots import make_subplots
import plotly.express as px
import plotly.graph_objects as go

# Create two separate subplots
fig = make_subplots(rows=1, cols=2, subplot_titles=['Entry Map', 'Exit Map'])

# Add the first map to the first subplot
fig.add_trace(
    go.Scattermapbox(
        lat=unified_entry["lat"],
        lon=unified_entry["lng"],
        mode='markers',
        hovertext=unified_entry["station_name"],
        marker=dict(
            size=unified_entry["n_passengers_entry"],
            color=unified_entry["n_passengers_entry"],
            colorscale=px.colors.cyclical.IceFire,
            colorbar=dict(title="Passengers"),
        ),
    ),
    row=1, col=1
)

# Set the layout for the first subplot
fig.update_mapboxes(style="carto-positron", zoom=9, row=1, col=1)

# Add the second map to the second subplot
fig.add_trace(
    go.Scattermapbox(
        lat=unified_exit["lat"],
        lon=unified_exit["lng"],
        mode='markers',
        hovertext=unified_exit["station_name"],
        marker=dict(
            size=unified_exit["n_passengers_exit"],
            color=unified_exit["n_passengers_exit"],
            colorscale=px.colors.cyclical.IceFire,
            colorbar=dict(title="Passengers"),
        ),
    ),
    row=1, col=2
)

# Set the layout for the second subplot
fig.update_mapboxes(style="open-street-map", zoom=9, row=1, col=2)

# Update the layout of the overall figure
fig.update_layout(height=600, width=1000, title_text="Side By Side Subplots")

# Show the figure
fig.show()
```
